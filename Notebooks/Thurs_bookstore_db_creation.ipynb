Use this markdown cell to write notes needed to design the new relational db schema

- transactions = ['transaction_id','store_id','transaction_date','member_id','book_id','quantity','discount_percent']

- members = ['member_id','member_first_name','member_last_name','member_email','member_tier','member_join_date','member_city','member_state'] 

- stores = ['store_id','store_name','store_city','store_state','store_region'] ***need store_id 

- authors = ['author_id','author_first','author_last'] ** need author_id

- publishers = ['publisher_id','publisher_name','publisher_city','publisher_country'] ** need publisher_id 

- books = ['book_id','isbn','title','genre','format','author_id','publication_year','publication_id','list_price'] ** author_id



```python
# filter the df into tables matching the schema in the markdown cell above

members_df = df[['member_id','member_first_name','member_last_name','member_email','member_tier','member_join_date','member_city','member_state']].drop_duplicates()
members_df.head()
```

```python
members_df.member_id.value_counts()
```

```python
stores_df = df[['store_name','store_city','store_state','store_region']].drop_duplicates().reset_index(drop=True)
stores_df
```

```python
stores_df.insert(0, 'store_id', range(1,len(stores_df)+1))
stores_df
```

```python
authors_df = df[['author_first','author_last']].drop_duplicates().reset_index(drop=True)
authors_df.insert(0, 'author_id', range(1,len(authors_df)+1))
authors_df.head()
```

```python
publishers_df = df[['publisher_name','publisher_city','publisher_country']].drop_duplicates().reset_index(drop=True)
publishers_df.insert(0, 'publisher_id', range(1,len(publishers_df)+1))
publishers_df.head()
```

```python
books_df = df[['book_id','isbn','title','genre','format','author_first','author_last','publication_year','publisher_name','list_price']].drop_duplicates().reset_index(drop=True)
books_df.head()
```

```python
books_df = books_df.merge(authors_df[['author_id','author_first','author_last']], left_on=['author_first','author_last'], right_on = ['author_first','author_last'], how='left')
books_df.head(3)
```

```python
books_df = books_df.drop(columns=['author_first','author_last'])
books_df.head()
```

```python
books_df = books_df.merge(publishers_df[['publisher_name', 'publisher_id']], on='publisher_name', how='left')
books_df.head()
```

```python
books_df = books_df.drop(columns='publisher_name')
books_df.head()
```

```python
transaction_df = df[['transaction_id','store_name','transaction_date','member_id','book_id','quantity','discount_percent']].drop_duplicates().reset_index(drop=True)
transaction_df.head()
```

```python
transaction_df = transaction_df.merge(stores_df[['store_id', 'store_name']], on='store_name', how='left')
transaction_df = transaction_df.drop(columns='store_name')
transaction_df.head()
```
